```python
import matplotlib.pyplot as plt
import math

i = range(-500, 500)
x = []
y = []
pa = 10
da = 0
for v in i:
    xx = v * 0.01
    a = (xx*xx - 2*xx - 7) * math.exp(xx)
    x.append(xx)
    y.append(a)
    pda = da # 이전 delta
    da = pa - a # 현재 delta
    pa = a
    if pda * da < 0:
        print("x", xx, "y", a, "dy", da)
plt.plot(x,y)
plt.show()
```
