```python
import matplotlib.pyplot as plt
import math
n = 100000
i = range(0, n)
x = []
y = []
s = 0
for v in i:
    xv = (math.log(2) - math.log(0.5)) * v / n +  math.log(0.5)
    yv = math.exp(2*xv)

    x.append(xv)
    y.append(yv)
    s += (math.log(2) - math.log(0.5)) / n * yv
print("area", s)
plt.plot(x,y)
plt.show()
```
