```python
import matplotlib.pyplot as plt
import math

n = 1000
i = range(1, n)
x = []
y = []
s = 5
for v in i:
    xv = v * 0.01
    s += (2 - 3 / xv**2) * 0.01
    yv = s
    x.append(xv)
    y.append(yv)

figure, axes = plt.subplots(1)
axes.set_aspect(1)
plt.plot(x,y)
plt.show()
```
