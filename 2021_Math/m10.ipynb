```python
import matplotlib.pyplot as plt
import math
n = 100
i = range(-n, n+1)
x = []
y = []
y1 = []
for v in i:
    xv = (1 / n) * v 
    yv = math.sqrt(1 - xv**2)
    x.append(xv)
    y.append(yv)
    y1.append(-yv)
    bl = math.sqrt( (xv+1)**2 + (yv)**2)
    print("B", xv, yv, bl)

figure, axes = plt.subplots(1)
axes.set_aspect(1)
plt.plot(x,y)
plt.plot(x, y1)
plt.show()
```
