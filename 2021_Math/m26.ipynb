```python
from itertools import permutations

a = [1,2,3,4,5,6]
permute = permutations(a, 6)
ps = list(permute)
print("전체 순열",len(ps))
print("원 순열", len(ps) / 6)
count = 0
for p in ps:
    for i in range(0,6):
        if p[i] == 1:
            if i == 0:
                l = p[5]
                ll = p[4]
            elif i == 1:
                l = p[0]
                ll = p[4]
            else:
                l = p[i-1]
                ll = p[i-2]
            if i == 5:
                r = p[0]
                rr = p[1]
            elif i == 4:
                r = p[5]
                rr = p[0]
            else:
                r = p[i+1]
                rr = p[i+2]
            if l == 2 and ll != 3:
                count += 1
                print(p)
            if r == 2 and rr != 3:
                count += 1
                print(p)
print("1과 2가 양 사이드 일 경우", count / 6)
```
